```python
import matplotlib.pyplot as plt
import numpy as np  

x=np.random.rand(1000)

plt.hist(x, bins=30, density=True, alpha=0.5, color='g', edgecolor='black')
plt.title('Histogram of Random Numbers')
plt.xlabel('Value')
plt.ylabel('Density')
plt.grid(axis='y', alpha=0.75)
plt.xticks(np.arange(0, 1.1, step=0.1))
plt.show()

```

Grafico de barras

```python
x=np.random.rand(100)

plt.bar(range(100), x, color='blue', alpha=0.7, edgecolor='black')
plt.title('Bar Chart of Random Numbers')    
```

Com figure

```python
x=[1, 2, 3, 4, 5]
y=[10, 7, 8, 6, 4]

fig=plt.figure()


fig, ax  = plt.subplots(figsize=(10, 5))
fig, ax2  = plt.subplots(figsize=(10, 5))
ax.plot(x, y, color='blue', marker='o', linestyle='-', linewidth=2, markersize=8)
ax2.plot(y, x, color='red', marker='o', linestyle='-', linewidth=2, markersize=8)
ax.set_title('Doenças Cardiovasculares')


# Definição do dicionário corrigida
prod = {'Product A': 150,
        'Product B': 200,
        'Product C': 300}

# Gráficos em subplots
fig, ax = plt.subplots(2, 2, figsize=(10, 10))

# Gráfico de barras
ax[0, 0].bar(prod.keys(), prod.values(), color='purple', alpha=0.7, edgecolor='black')
ax[0, 0].set_title('Gráfico de Barras')

# Gráfico de barras horizontais
ax[0, 1].barh(prod.keys(), prod.values(), color='purple', alpha=0.7, edgecolor='black')
ax[0, 1].set_title('Gráfico de Barras Horizontais')

# Gráfico de dispersão
scatter = ax[1, 0].scatter(prod.keys(), prod.values(), color='purple', alpha=0.7, edgecolor='black')
ax[1, 0].set_title('Gráfico de Dispersão')

# Adicionar legenda ao gráfico de dispersão
ax[1, 0].legend(*scatter.legend_elements(), loc="upper right", title="Classes")

# Ajustar layout
plt.tight_layout()
plt.show()

```
