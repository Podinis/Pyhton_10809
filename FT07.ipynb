# üìò Visualiza√ß√£o de Dados com Seaborn (em Python)

## üîé O que √© o Seaborn?

O **Seaborn** √© uma biblioteca de visualiza√ß√£o de dados baseada no Matplotlib que fornece uma interface de alto n√≠vel para a cria√ß√£o de gr√°ficos estat√≠sticos informativos e esteticamente agrad√°veis.

- Facilita a cria√ß√£o de **gr√°ficos complexos** com poucas linhas de c√≥digo.
- Est√° profundamente integrada com o `pandas`, permitindo trabalhar com `DataFrames` de forma eficiente.
- Suporta **gr√°ficos estat√≠sticos**, como gr√°ficos de regress√£o, distribui√ß√£o, boxplots, mapas de calor, entre outros.

üìö Documenta√ß√£o oficial: [https://seaborn.pydata.org](https://seaborn.pydata.org)


````python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Carregar um dataset limpo: 'penguins'
df = sns.load_dataset("penguins")
df.dropna(inplace=True)  # Remover valores nulos
df.head()
````

## üé® Est√©tica e Estilo dos Gr√°ficos com Seaborn

O Seaborn permite aplicar estilos e cores predefinidos para tornar os gr√°ficos mais apelativos.

```python
sns.set_style("whitegrid")  # Outros estilos: darkgrid, white, ticks, dark
sns.set_palette("pastel")   # Outros: deep, dark, colorblind, bright
```

Estes comandos podem ser aplicados no in√≠cio do notebook para afetar todos os gr√°ficos seguintes.


## üìä Gr√°ficos de Dispers√£o (`scatterplot`)

Usado para visualizar a rela√ß√£o entre duas vari√°veis num√©ricas.


````python
# Criamos um gr√°fico de dispers√£o com as vari√°veis 'flipper_length_mm' no eixo X e 'body_mass_g' no eixo Y. Os pontos s√£o coloridos por esp√©cie.
sns.scatterplot(data=df, x="flipper_length_mm", y="body_mass_g", hue="species")
# Adicionamos um t√≠tulo ao gr√°fico.
plt.title("Comprimento da Barbatana vs Massa Corporal")
# Definimos o r√≥tulo do eixo X.
plt.xlabel("Comprimento da Barbatana (mm)")
# Definimos o r√≥tulo do eixo Y.
plt.ylabel("Massa Corporal (g)")
# Mostramos o gr√°fico.
plt.show()
````

## üìà Gr√°ficos de Linhas (`lineplot`)

√ötil para dados sequenciais ou varia√ß√µes m√©dias.


````python
# üìç Explica√ß√£o:
# - `sns.lineplot()` cria um gr√°fico de linhas.
# - Ideal para mostrar tend√™ncias, mesmo em dados n√£o temporais.
# - Representamos a massa corporal m√©dia em fun√ß√£o do comprimento da barbatana, agrupado por esp√©cie.

sns.lineplot(data=df, x="flipper_length_mm", y="body_mass_g", hue="species")
plt.title("Tend√™ncia do Comprimento da Barbatana e Massa Corporal")
plt.xlabel("Comprimento da Barbatana (mm)")
plt.ylabel("Massa Corporal (g)")
plt.show()
````

## üìä Histogramas e Densidade (`histplot` e `kdeplot`)

Visualizam a distribui√ß√£o de uma vari√°vel cont√≠nua.


````python
# Criamos um histograma da vari√°vel 'flipper_length_mm', com curva de densidade (kde). Os dados s√£o separados por esp√©cie.
sns.histplot(data=df, x="flipper_length_mm", kde=True, hue="species", multiple="stack")
# Adicionamos um t√≠tulo ao gr√°fico.
plt.title("Distribui√ß√£o do Comprimento da Barbatana por Esp√©cie")
# Definimos o r√≥tulo do eixo X.
plt.xlabel("Comprimento da Barbatana (mm)")
# Definimos o r√≥tulo do eixo Y.
plt.ylabel("Frequ√™ncia")
# Mostramos o gr√°fico.
plt.show()
````

````python
# Criamos um histograma da vari√°vel 'flipper_length_mm', com curva de densidade (kde). Os dados s√£o separados por esp√©cie.
sns.kdeplot(data=df, x="flipper_length_mm", hue="species", fill=True, common_norm=False, alpha=0.5)
# Adicionamos um t√≠tulo ao gr√°fico.
plt.title("Distribui√ß√£o do Comprimento da Barbatana por Esp√©cie")
# Definimos o r√≥tulo do eixo X.
plt.xlabel("Comprimento da Barbatana (mm)")
# Definimos o r√≥tulo do eixo Y.
plt.ylabel("Frequ√™ncia")
# Mostramos o gr√°fico.
plt.show()
````

## üß™ Gr√°ficos de Caixa e Violino (`boxplot` e `violinplot`)

√öteis para visualizar distribui√ß√£o, mediana e outliers.


````python
# Criamos um gr√°fico de caixa (boxplot) com as esp√©cies no eixo X e a massa corporal no eixo Y.
sns.boxplot(data=df, x="species", y="body_mass_g")
# Adicionamos um t√≠tulo ao gr√°fico.
plt.title("Boxplot da Massa Corporal por Esp√©cie")
# Definimos o r√≥tulo do eixo X.
plt.xlabel("Esp√©cie")
# Definimos o r√≥tulo do eixo Y.
plt.ylabel("Massa Corporal (g)")
# Mostramos o gr√°fico.
plt.show()
````

````python
# Criamos um gr√°fico de caixa (boxplot) com as esp√©cies no eixo X e a massa corporal no eixo Y.
sns.violinplot(data=df, x="species", y="body_mass_g")
# Adicionamos um t√≠tulo ao gr√°fico.
plt.title("Violinplot da Massa Corporal por Esp√©cie")
# Definimos o r√≥tulo do eixo X.
plt.xlabel("Esp√©cie")
# Definimos o r√≥tulo do eixo Y.
plt.ylabel("Massa Corporal (g)")
# Mostramos o gr√°fico.
plt.show()
````

## üìä Gr√°ficos de Barras (`barplot`)

Mostra estat√≠sticas agregadas (por padr√£o a m√©dia).


````python
# Criamos um gr√°fico de barras com a massa corporal m√©dia por ilha.
sns.barplot(data=df, x="island", y="body_mass_g")
# Adicionamos o t√≠tulo ao gr√°fico.
plt.title("Massa Corporal M√©dia por Ilha")
# Definimos o r√≥tulo do eixo X.
plt.xlabel("Ilha")
# Definimos o r√≥tulo do eixo Y.
plt.ylabel("Massa Corporal (g)")
# Mostramos o gr√°fico.
plt.show()
````

## üå°Ô∏è Mapas de Calor (`heatmap`)

Visualiza√ß√£o de correla√ß√µes ou matrizes de dados.


````python
# Calculamos a matriz de correla√ß√£o entre as colunas num√©ricas do dataset.
correlacoes = df.corr(numeric_only=True)
# Criamos o mapa de calor (heatmap), com os valores anotados e uma paleta de cores 'coolwarm'.
sns.heatmap(correlacoes, annot=True, cmap="coolwarm")
# Adicionamos o t√≠tulo ao gr√°fico.
plt.title("Mapa de Calor das Correla√ß√µes")
# Mostramos o gr√°fico.
plt.show()
````

## üîó Gr√°ficos de Pares (`pairplot`)

Permite visualizar todas as combina√ß√µes entre m√∫ltiplas vari√°veis num√©ricas.


````python
# Criamos uma matriz de gr√°ficos de dispers√£o entre todas as vari√°veis num√©ricas, diferenciando por esp√©cie.
sns.pairplot(df, hue="species")
# Adicionamos um t√≠tulo geral √† figura com deslocamento vertical.
plt.suptitle("Gr√°ficos de Pares das Vari√°veis Num√©ricas", y=1.02)
# Mostramos o gr√°fico.
plt.show()
````

## üé® Controlo da Est√©tica com Seaborn

O Seaborn permite personalizar a apar√™ncia geral dos gr√°ficos atrav√©s de quatro componentes principais:

1. **Estilos de fundo (`set_style`)** ‚Äì Define o fundo do gr√°fico.
2. **Paletas de cores (`set_palette`)** ‚Äì Define as cores dos elementos.
3. **Contexto visual (`set_context`)** ‚Äì Adapta o tamanho dos elementos para diferentes usos.
4. **Tamanho da figura (`plt.figure(figsize=(x, y))`)** ‚Äì Define o tamanho da imagem final.



### 1Ô∏è‚É£ Estilos de Fundo

Os estilos dispon√≠veis s√£o:

- `"white"`
- `"dark"`
- `"whitegrid"` (√∫til para dados com grelhas)
- `"darkgrid"`
- `"ticks"` (com marcas subtis nos eixos)

```python
sns.set_style("whitegrid")
```


### 2Ô∏è‚É£ Paletas de Cores

Paletas predefinidas para uma est√©tica mais consistente:

- `"deep"` (predefinido)
- `"muted"`
- `"pastel"`
- `"bright"`
- `"dark"`
- `"colorblind"`

```python
sns.set_palette("pastel")
```


### 3Ô∏è‚É£ Contexto Visual

Ajusta automaticamente o tamanho dos elementos visuais (fontes, pontos, linhas):

- `"paper"` (ideal para impress√£o)
- `"notebook"` (predefinido)
- `"talk"` (para apresenta√ß√µes)
- `"poster"` (para gr√°ficos grandes)

```python
sns.set_context("talk")
```

````python
# Criamos um gr√°fico de caixa (boxplot) com as esp√©cies no eixo X e a massa corporal no eixo Y.
sns.boxplot(data=df, x="species", y="body_mass_g")
# Adicionamos um t√≠tulo ao gr√°fico.
plt.title("Boxplot da Massa Corporal por Esp√©cie")
# Definimos o r√≥tulo do eixo X.
plt.xlabel("Esp√©cie")
# Definimos o r√≥tulo do eixo Y.
plt.ylabel("Massa Corporal (g)")
# Mostramos o gr√°fico.
plt.show()
````

## üß† Lista de Exerc√≠cios com Seaborn

1. Cria um gr√°fico de dispers√£o entre o comprimento e a massa corporal, diferenciando as esp√©cies com cores.
2. Representa a distribui√ß√£o do comprimento da barbatana com `histplot`, incluindo `kde=True`.
3. Compara a massa corporal entre esp√©cies com um `boxplot`.
4. Cria um `violinplot` para a vari√°vel `flipper_length_mm`.
5. Constr√≥i um `heatmap` das correla√ß√µes entre as vari√°veis num√©ricas.
6. Usa o `pairplot` para estudar a rela√ß√£o entre todas as vari√°veis num√©ricas do dataset.
7. Cria um `barplot` que mostre a m√©dia da massa corporal por ilha.

üìù Utiliza apenas `seaborn` para resolver os exerc√≠cios.


````python
# 1 Grafico dispers√£o com as vari√°veis 'flipper_length_mm' no eixo X e 'body_mass_g' no eixo Y. Os pontos s√£o coloridos por esp√©cie.
sns.scatterplot(data=df, x="flipper_length_mm", y="body_mass_g", hue="species", style="species", palette="dark") # o styles muda o tipo de ponto
plt.title("Comprimento da Barbatana vs Massa Corporal")
plt.xlabel("Comprimento da Barbatana (mm)")
plt.ylabel("Massa Corporal (g)")
plt.legend(title="Esp√©cie", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
````

````python
# 2 - Grafico histplot com a vari√°vel 'flipper_length_mm', com curva de densidade (kde). Os dados s√£o separados por esp√©cie.
sns.histplot(data=df, x="flipper_length_mm", kde=True, hue="species", multiple="stack",shrink=.8) #Se colocar pallete="dark" fica mais escuro, shrink √© o espa√ßamento
plt.title("Distribui√ß√£o do Comprimento da Barbatana por Esp√©cie")
plt.xlabel("Comprimento da Barbatana (mm)")
plt.ylabel("Frequ√™ncia")
plt.show()
````

````python
# 3 - Criar um grafico boxplot com as esp√©cies no eixo X e a massa corporal no eixo Y.
sns.boxplot(data=df, x="species", y="body_mass_g", gap=.2) 
plt.title("Boxplot da Massa Corporal por Esp√©cie")
plt.xlabel("Esp√©cie")
plt.ylabel("Massa Corporal (g)")
plt.show()
````

````python
# 4 - Criar um grafico violinplot com as esp√©cies no eixo X e flipper_length_mm no eixo Y.
sns.violinplot(data=df, x="species", y="flipper_length_mm",legend=True)
plt.title("Violinplot do Comprimento da Barbatana por Esp√©cie")
plt.xlabel("Esp√©cie")
plt.ylabel("Comprimento da Barbatana (mm)")
plt.show()
````

````python
# 4 - Criar um grafico violinplot com as esp√©cies no eixo X e flipper_length_mm no eixo Y.
sns.violinplot(data=df, y="flipper_length_mm",hue="species",legend=True)
plt.title("Violinplot do Comprimento da Barbatana por Esp√©cie")
plt.xlabel("Esp√©cie")
plt.ylabel("Comprimento da Barbatana (mm)")
plt.show()
````

````python
# 5 - Criar um grafico heatmap com a matriz de correla√ß√£o entre as colunas num√©ricas do dataset.
correlacoes = df.corr(numeric_only=True)
sns.heatmap(correlacoes,vmin=-1,vmax=1, annot=True, cmap="coolwarm")
plt.title("Mapa de Calor das Correla√ß√µes")
plt.xlabel("Vari√°veis")
plt.show()
````

````python
# 6 - Criar um grafico pairplot entre todas as vari√°veis num√©ricas do dataset

sns.pairplot(df, hue="species")
plt.suptitle("Gr√°ficos de Pares das Vari√°veis Num√©ricas", y=1.05) #o y corresponde a distancia a partir do topo  
plt.show()
      
````

````python
# 7 - Criar um grafico de barras com a massa corporal m√©dia por ilha.

sns.barplot(data=df, x="island", y="body_mass_g") # o estimator √© a fun√ß√£o que calcula a m√©dia, neste caso a m√©dia aritm√©tica
# Adicionamos o t√≠tulo ao gr√°fico.
plt.title("Massa Corporal M√©dia por Ilha")
# Definimos o r√≥tulo do eixo X.
plt.xlabel("Ilha", fontsize=12, fontweight='bold')
# Definimos o r√≥tulo do eixo Y.
plt.ylabel("Massa Corporal (g)")
# Mostramos o gr√°fico.
plt.show()
````
